#### 4. Import the users table.

```python
users=pd.read_sql_table("users", engine)
```

```python
users.head()
```

#### 5. Rename Id column to userId.

```python
users=users.rename(columns={'Id':'userId'})
users.head()
```

#### 6. Import the posts table. 

```python
posts=pd.read_sql_table("posts", engine)
posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts=posts.rename(columns={"Id":"postId", "OwnerUserId":"userId"})
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_new=users[["userId", "Reputation", "Views", "UpVotes", "DownVotes"]]
users_new
```

```python
posts_new=posts[["postId", "Score", "userId", "ViewCount", "CommentCount"]]
posts_new
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
df=users_new.merge(posts_new, on="userId")
df.head()
```
