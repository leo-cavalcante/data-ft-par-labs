# Data Cleaning

## Functions for droping columns that don't have significant value

```python
def jobs_drop_columns(df):
    """
    This function automates the identification of columns that don't interest us and drops them all
    
    Input: the initial JOBS database
    Output: the final JOBS database, with less columns
    """
    
    # Droping columns that present all NaN values:
    df.dropna(axis=1, how="all", inplace=True)
    
    # Initializing the variable that will store the columns' names to be droped
    columns_to_drop=[]
    
    # Droping all columns that present at least 80% of NaN values:
    limit = 0.3 * df.shape[0]
    
    for col in df.columns:
        if df[col].isna().sum()>=limit:
            columns_to_drop.append(str(col))
    
    # Droping columns that are not in English language (".en")
    for language in [".cs", ".es", ".fr", ".sk"]:
        columns_to_drop += [c for c in df.columns if language in c]
        
    # Droping all useless columns that were manually identified as useless
    for text in ["_highlightResult.", "_image.", "logo."]:
        columns_to_drop += [c for c in df.columns if text in c]
    
    # Cleaning all duplicates from columns list:
    columns_to_drop = list(set(columns_to_drop))
    
    # Finally, drop all columns previously selected
    df.drop(columns=columns_to_drop, inplace=True)
    
    return df
```

## Function to drop additional columns (previously analyzed)

```python
def jobs_drop_remaining_columns(df):
    """
    This function drop columns by their full names, which have been identified (through analysis)
    to be useless for our objectives here.
    
    Input: the initial JOBS database
    Output: the final JOBS database, with less columns
    """
    
    # Explanation:
    ## 'organization.slug' is the slug of the ancient name of the company (some time ago) --> more than 80% of values are duplicated from 'organization.website_organization.slug'
    ## 'profession_name.en.Tech' is a subset of 'profession.name.en'
    ## 'website.reference' has the same value ("wttf") for all lines, indicating the scraping source (Welcome to the Jungle).
    ## 'contract_type_names.en' has the same value as 'contract_type'
    ## 'sectors_name.en' is a list of dictionaries also found at "'sectors" column
    
    columns_to_drop = []
    
    for col in df.columns:
        if col in ["organization.slug",'profession_name.en.Tech', 'website.reference', 'contract_type_names.en', 'sectors_name.en']:
            columns_to_drop.append(col)
    df.drop(columns=columns_to_drop, inplace=True)
    
    return df
```

## Cleaning values

### Sectors: unpacking values

This column is composed of dictionaries containing the sectors parents and the sectors.

Both segments interest me and both can have more than 1 value.

Therefore, it will be to unpack these values in a new column and, finally, 

```python
print(jobs["sectors"])
print(jobs["sectors"][3])
print(jobs["sectors"][3][0])
print(jobs["sectors"][3][0]["name"])
print(jobs["sectors"][3][0]["name"]["en"])
```

As we can see, the list may have one or more sectors for each job/company.

Also, it may have one parent sector or more for each.

Both informations interest me, so I will unpack them in 2 different columns.

```python
def clean_sectors(sectors):
    return " / ".join([s["name"]["en"] for s in sectors])

def clean_sectors_parents(sectors):
    return " / ".join(list(set([s["parent"]["en"] for s in sectors])))
```

### Profile cleaning

```python
from bs4 import BeautifulSoup
from markdown import markdown
import re

def markdown_to_text(markdown_string):
    """ Converts a markdown string to plaintext """

    # md -> html -> text since BeautifulSoup can extract text cleanly
    html = markdown(markdown_string)

    # remove code snippets
    html = re.sub(r'<pre>(.*?)</pre>', ' ', html)
    html = re.sub(r'<code>(.*?)</code >', ' ', html)

    # extract text
    soup = BeautifulSoup(html, "html.parser")
    text = ''.join(soup.findAll(text=True))

    return text
```

### Cleaning it all

```python
def jobs_formatting(df):
    """
    This function effectuates the necessary cleaning for columns that contains lists, dictionaries, types that are not
    recognized as such, eliminating tags from html and put data in columns in a better format.
    
    Input: dataframe to be cleaned
    Output: dataframe with clean columns
    """
    
    # Creating a column for the sectors_parents, from unpacking the values from the "sectors" column
    df["sectors_parents"] = df["sectors"].apply(clean_sectors_parents)
    
    # Replacing the values in the "sectors" column by its values unpacked from the dictionary
    df["sectors"] = df["sectors"].apply(clean_sectors)
    
    # Eliminating all html tags from the text, by applying the markdown_to_text function stated before
    df["profile"]=df["profile"].fillna('').apply(markdown_to_text)
    
    # Change the type of the column, so it will be considered as DateTime as it should
    df["published_at"] = df["published_at"].astype("datetime64")
    
    # Making the column "language" uppercase
    df["language"] = df["language"].apply(lambda x: x.upper())

    # Making the column "contract-type" capitalized
    df["contract_type"] = df["contract_type"].str.replace("_","-")
    df["contract_type"] = df["contract_type"].apply(lambda x: x.capitalize())
    
    return df
```

# Extracting data from the companies webpage in Welcome to the Jungle

Now that we got the data from the jobs page, it is possible to take all the companies appeared in our search, visit their webpage and extract their data. That's what will be done in this section.

```python
from bs4 import BeautifulSoup

def companies_scrap_data(df):
    """This function captures the companies' slug, visit the webpage of each slug and returns
    all relevant informations in the format of a database"""
    
    # Getting all the URLs in a list format
    url_list = list(set([(lambda x: "https://www.welcometothejungle.com/en/companies/" + x)(x) for x in df["organization.website_organization.slug"]]))
    
    # Getting the headers that will be common to all websites
    text = """:authority: client.axept.io
:method: GET
:path: /5f1aa8bf72fdae0c1a7a49dc.json
:scheme: https
accept: */*
accept-encoding: gzip, deflate, br
accept-language: en-GB,en-US;q=0.9,en;q=0.8
cache-control: no-cache
origin: https://www.welcometothejungle.com
pragma: no-cache
referer: https://www.welcometothejungle.com/en/companies/getaround
sec-fetch-dest: empty
sec-fetch-mode: cors
sec-fetch-site: cross-site
user-agent: Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36"""
    
    # Initialize our dataframe that whill be returned
    companies = pd.DataFrame()
    
    # Looping through every company webpage on Welcome to the Jungle
    for url in url_list:
        
        # Initializing the temporary dataframe that will collect the data scraped
        temp = pd.DataFrame()
        
        # Getting the correct URL and soup
        html = requests.get(url).content
        soup = BeautifulSoup(html)
        
        # SCRAPING DATA FROM THE SOUP
        print(str(url[48:]), end="  ")
        
        temp['Company Name'] = [i.text for i in soup.select('h1')]
        
        if len(soup.select('[class="q4qgh9-9 bnTGzJ"]'))>0:
            temp['Open Jobs'] = soup.select('[class="q4qgh9-9 bnTGzJ"]')[0]
        
        if len(soup.select('span[class="sc-1qc42fc-2 dJqCnn"]'))>0:
            temp['Sectors'] = [i.text for i in soup.select('span[class="sc-1qc42fc-2 dJqCnn"]')]
        
        if len(soup.select("a.sc-oUcyK"))>0:
            temp['Website'] = soup.select("a.sc-oUcyK")[0]["href"]
        
        if len(soup.select('p.sc-1yjljb7-4'))>0:
            temp['Street'] = " / ".join([i.text for i in soup.select('p.sc-1yjljb7-4')])
        
        if len(soup.select('p.sc-1yjljb7-5>span:first-child'))>0:
            temp['Postal Code'] = " / ".join([i.text for i in soup.select('p.sc-1yjljb7-5>span:first-child')])
        
        if len(soup.select('p.sc-1yjljb7-5>span:last-child'))>0:
            temp['City'] = " / ".join([i.text for i in soup.select('p.sc-1yjljb7-5>span:last-child')])
        
        if len(soup.select('span.sc-1qc42fc-2'))>1:
            temp['All Cities'] = soup.select('span.sc-1qc42fc-2')[1]
        
            ## Getting almost all the data from the unordered list situated in the page, except "Parité"
            ## Data = Year Founded, #Employees, Turnover, Chiffre d'Affaires, etc.
        lst = soup.findAll("li",{"class":"sc-1n18lhk-1 bCWtWO"})
        for index, name in enumerate(lst):
            temp[[i.text for item in lst for i in item.select("h4")][index]] = [i for i in name.select('[class="sc-1n18lhk-3 iQuKws"]')[0]][0]

            ## Getting the data left from the unordered list situated in the page (if they exist): "Parité"
        lst = soup.findAll("li",{"class":"sc-1n18lhk-1 eoxRwK"})
        if len(lst) >=1:
            for index, name in enumerate(lst):
                temp[[i.text+(" (F, M)") for item in lst for i in item.select("h4")][index]] = ", ".join([i.text for i in name.select('[class="sc-1n18lhk-3 iQuKws"]')])

            ## Getting the data from the text blocks at the right margin, explaining things about the company
        lst_1 = [i.text for i in soup.select('header[class="sc-2zd334-0 fKWrMT"]')]
        lst_2 = [i.text for i in soup.select('header[class="sc-2zd334-0 fKWrMT"]+div')]
            
        for index in range(len(lst_1)):
            if lst_1!="":
                temp[lst_1[index]] = lst_2[index]
        
        companies = pd.concat([companies,temp], axis=0)
    
    print("***** SCRAPING FINISHED *****")
    
    return companies
```
