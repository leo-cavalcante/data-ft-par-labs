## Cleaning the column "Fatal (Y/N)"

```python
sharks["Fatal (Y/N)"].value_counts()
```

```python
sharks["Fatal (Y/N)"]=sharks["Fatal (Y/N)"].str.replace("N ","N")
```

```python
sharks["Fatal (Y/N)"]=sharks["Fatal (Y/N)"].str.replace("N ","N")
```

```python
sharks["Fatal (Y/N)"].value_counts()
```

```python
sharks["Fatal (Y/N)"]=sharks["Fatal (Y/N)"].str.replace("y","Y").replace("M","UNKNOWN").replace("2017","UNKNOWN")
```

```python
sharks["Fatal (Y/N)"].value_counts()
```

```python
sharks["Fatal (Y/N)"]=sharks["Fatal (Y/N)"].fillna(value="UNKNOWN")
```
