### Columns "href" and "href formula"

```python
(sharks["href"]==sharks["href formula"]).sum()
```

```python
link=sharks[["href","href formula"]]
link
```

```python
link[link["href"]!=link["href formula"]]
```

```python
indexes = list(link[link["href"]!=link["href formula"]].index)
indexes
```

```python
# Checking the href that are valid

import requests

for index in indexes:
    #print(link.loc[index,"href"])
    try:
        request = requests.get(link.loc[index,"href"])
    except:
        link.loc[index,"href"]=0
    if request.status_code == 200:
        pass
    else:
        link.loc[index,"href"]=0
        
link[link["href"]!=link["href formula"]]
```

We check that only one line (with index 587) has an invalid URL. Let's then attribute the value of href:

```python
# Checking the attribution first
link.loc[587,"href"]=link.loc[587,"href formula"]
link[link["href"]!=link["href formula"]]
```

```python
# Attributing the value at last
sharks.loc[587,"href"]=sharks.loc[587,"href formula"]
```
