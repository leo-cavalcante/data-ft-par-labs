## IMDB Top 1000

```python
url='https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=101&ref_=adv_nxt'
```

```python
soup=BeautifulSoup(r.get(url).content)
```

```python
len(soup.select('div.lister-item.mode-advanced'))
```

```python
selection=soup.select('div.lister-item.mode-advanced')
```

```python
selection[0].select('h3>a')
```

```python
selection[0].select('div.lister-item.mode-advanced span.runtime')
```

```python
titles=[i.select('h3>a')[0].text for i in selection]
```

```python
runtime=[int(i.select('span.runtime')[0].text.split()[0]) for i in selection]
```

```python
metascores=[i.select('span.metascore') for i in selection]
```

```python
metascores=[-1 if len(i)==0 else int(i[0].text.strip()) for i in metascores]
```

```python
import pandas as pd
```

```python
pd.DataFrame([titles, runtime, metascores]).T
```

```python
import time
dftotal=pd.DataFrame()
for i in range(10):
    url=f'https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start={1+100*(i)}&ref_=adv_nxt'
    soup=BeautifulSoup(r.get(url).content)    
    selection=soup.select('div.lister-item.mode-advanced')    
    titles=[i.select('h3>a')[0].text for i in selection]
    runtime=[int(i.select('span.runtime')[0].text.split()[0]) for i in selection]
    metascores=[i.select('span.metascore') for i in selection]
    metascores=[-1 if len(i)==0 else int(i[0].text.strip()) for i in metascores]
    df=pd.DataFrame([titles, runtime, metascores]).T
    dftotal=dftotal.append(df)
    time.sleep(2)
    print(i, 'is finished')
```

```python
dftotal
```

In order to get the data from RESTful API based website we follow the following steps:
1. Open the url
2. Go to the secong page (usually it allows to identify how the pagination works, since on the first page you dont have page=1)
3. Grab the link (and request headers) to python
4. send request
5. get html content and convert it into a soup
6. detect the correct selectors for every datapoint you need to select. If data is located 1 level deeper, simply collect all the links on this stage.  
6b. Loop over list of links if needed and detect the correct selectors for every datapoint you need to select.
7. Merge everything into a big dataframe. PROFIT!
