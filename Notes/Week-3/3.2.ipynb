In order to get the data from RESTful API based website we follow the following steps:
1. Open the url
2. Inspect the page in the network tab
3. Go to the secong page (usually it allows to identify how the pagination works, since on the first page you dont have page=1)
4. Search for the link that will lead you to data
5. Grab the link and request headers to python
6. send request
7. get the json from the response
8. normalize it and store into a df
9. loop over a bunch of pages and append the result to df.
10. PROFIT!

Workflow with RESTful API based websites:
1. Get the url to the website
2. Locate the data 
3. Get the url, headers (and payload if post request) to the API
4. Figure out how the pagination works. It can look like page=3 ore it can be offset=64 or any other keyword that changes in url (or payload for post)
5. make a request to get data for 1 single page. 
6. if failed - check the headers, payload. go to another page, etc. 
7. if still failed - try to get the data directly without api
8. if still failed - cry. there is nothing else to do. noone is going to help you on stackoverflow, because noone knows how to deal with your problem...
9. use selenium.
