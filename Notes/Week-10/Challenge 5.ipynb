```python
import pandas as pd
import numpy as np
```

```python
adults = pd.read_csv("adult.data",names=['age','employer','number_1','education','number_2','married_status',
                                         'profession','family_status','race','gender','number_3','number_4','hours_per_week',
                                        'country','salary'])
adults
```

```python
# Sex repartition
adults.gender.value_counts()
```

```python
# Female average age
adults[adults.gender==' Female'].age.mean()
```

```python
# percentage of German citizens 
adults.country.value_counts() / adults.shape[0]
```

```python
adults.salary.value_counts()
```

```python
# Mean and stddev of age of those that earn MORE than 50k€/year
#adults.salary.value_counts().index
adults[adults.salary==' >50K'].age.agg(['mean','std'])
```

```python
# Mean and stddev of age of those that earn LESS than 50k€/year
#adults.salary.value_counts().index
adults[adults.salary==' <=50K'].age.agg(['mean','std'])
```

```python
# Is it true that people that earn more than 50K€ have at least High School Education ?
print(adults[adults.salary==' >50K'].education.unique())
adults[adults.salary==' >50K'].education.value_counts()
```

```python
# Maximum age of Amer-Indian-Eskimo race
adults[adults.race==' Amer-Indian-Eskimo'].age.max()
```

```python
adults[adults.salary==' >50K'].married_status.value_counts()
```

```python
# Who has the most proportion of people earning more than 50k ? Married or bachelors ?

df = adults[["married_status","salary"]]
df["married?"] = df["married_status"].apply(lambda x: "Yes" if "Married-" in x else "No")

pd.crosstab(df["married?"], df.salary,margins=True, normalize='index')
```

```python
# Maximum amount of hours of work per week
print("Maximum amount of hours of work per week:",adults.hours_per_week.max())

# Number of people:
print("Number of people working lmaximum amount of hours of work per week:",
      adults[adults.hours_per_week ==adults.hours_per_week.max()].shape[0])

# Percentage of rich:
df = adults[adults.hours_per_week ==adults.hours_per_week.max()]
percent_rich = (df.salary.value_counts()/df.shape[0])[1]
print("Number of people working lmaximum amount of hours of work per week:",
      (percent_rich*100).round(2),"%")
```
